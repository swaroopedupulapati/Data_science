```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
file=pd.read_csv('energy.csv')
print(file)
#line chart ap kar
a=file[file['State']=='Andhra']
b=file[file['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].plot(a['Date'],a['Consumption'],marker='*',color="blue")
diag[1].plot(b['Date'],b['Consumption'],marker='_',color="orange")

a=file[file['State']=='Telengana']
b=file[file['State']=='Tripura']
fig,diag=plt.subplots(2)
diag[0].bar(a['Date'],a['Consumption'],color="skyblue")
diag[1].bar(b['Date'],b['Consumption'],color="olive")

cse4='Kerala'
cse5='Delhi'
t=[cse4,cse5]
a=file[file['State']=='Kerala']['Consumption'].sum()
b=file[file['State']=='Delhi']['Consumption'].sum()
y=[a,b]

fig,diag=plt.subplots(2)
diag[0].pie(y,labels=t)
```
